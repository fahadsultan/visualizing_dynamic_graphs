```python
import csv
header = ['origin', 'destination', 'count']

cluster_1 = ['JFK', 'ATL','ORD', 'LAX']  

cluster_2 = ['DFW', 'CLT', 'PHX', 'PHL']

n_time = 11

f = open('myData.csv', 'w')

writer = csv.writer(f)

writer.writerow(header)    
        
def create_pattern_1(f,airport_list):
    #all to all
    count = 0
    t= 0
    while t<= n_time:
        i = 0
        
        while i < len(airport_list):
            j = 0
            while j < len(airport_list):
                f.write('%s_%s,%s_%s,%s\n' %(airport_list[i],t,airport_list[j],t+1,t))
                
                j = j + 1
     
            i = i + 1
        t= t+1

def create_pattern_2(f,airport_list):
    #create pattern 2 is 1-1
    t= 0
    while t<= n_time:
        i = 0
        while i < len(airport_list):
            
            
            f.write('%s_%s,%s_%s,1\n' %(airport_list[i],t,airport_list[i],t+1))
                
            i = i + 1
        t= t+1

def create_pattern_3(f,airport_list):
    #create pattern 2 is 1 - all.
    #Doesnt look very good
    t= 0
    while t<= n_time:
        i = 0
        while i < len(airport_list):
            
            f.write('%s_%s,%s_%s,1\n' %(airport_list[0],t,airport_list[i],t+1))
            i = i + 1
        t= t+1


create_pattern_1(f, cluster_1)
create_pattern_1(f, cluster_2)

# writer.writerow('lax_5', DFW_)

f.close()
```
